```python
import pyspark.sql.functions as F

current_snapshot_df = spark.read.table(
    "contraloria.employee_payroll.silver_employee_payroll_latest_snapshot"
).select(
    "composite_key",
    "first_name",
    "last_name",
    "id_number",
    "salary",
    "allowance",
    (F.col('allowance') + F.col('salary')).alias('total_compensation'),
    "status_sp",
    "institution_sp",
    "position_sp",
    "start_date",
    "years_of_service",
)

current_snapshot_df.select("total_compensation").summary().show()
```

```python
current_snapshot_with_log_df = current_snapshot_df.withColumn(
    "total_compensation_log",
    F.log1p(F.col("total_compensation"))
)


current_snapshot_with_log_df.select(
    F.skewness("total_compensation").alias("skew_original"),
    F.skewness("total_compensation_log").alias("skew_log")
).show()
```

```python
current_snapshot_with_log_df.plot.hist('total_compensation_log',bins=100)
```
