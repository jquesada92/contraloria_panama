```sql
CREATE OR REPLACE VIEW  contraloria.reference_and_audit.v_api_check_log_last_run as
SELECT
  *,
  concat(institution_name_spanish, ' - ', status_name_spanish) as report,
  case
    when run_status = 'OK' then 1
    else 0
  end as OK,
  case
    when run_status <> 'OK' then 1
    else 0
  end as ERROR,
  max(checked_at) OVER (PARTITION BY institution_name_spanish, status_name_spanish) as last_run
FROM
  contraloria.reference_and_audit.api_check_log
qualify
  last_run = checked_at
```
